```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Series

```python
n_muestras = 100
s_blanca_X  = pd.Series(np.random.randn(n_muestras))
paseo_X     = s_blanca_X.cumsum()
s_blanca_Y  = pd.Series(np.random.randn(n_muestras))
paseo_Y     = s_blanca_Y.cumsum()
```

```python
plt.figure(figsize=(12,10))
plt.subplot(211)
s_blanca_X.plot()
plt.xlim(0,n_muestras)
plt.xlabel('Muestras')
plt.title('Serie blanca - X')
plt.subplot(212)
s_blanca_Y.plot()
plt.xlim(0,n_muestras)
plt.xlabel('Muestras')
plt.title('Serie blanca - Y')
```

```python
plt.subplot(211)
paseo_X.plot(figsize=(12,6),xlim=(-1,n_muestras),title='Paseo aleatorio - X')
plt.subplot(212)
paseo_Y.plot(figsize=(12,6),xlim=(-1,n_muestras),title='Paseo aleatorio - Y')
```

```python
print(paseo_X[0], paseo_Y[0])
print(paseo_X.at[0], paseo_Y.at[0])
```

```python
plt.figure(figsize=(12,12))
plt.plot(paseo_X.values,paseo_Y.values)
plt.plot(paseo_X.values[0],paseo_Y.values[0],'ro')
plt.plot(paseo_X.values[-1],paseo_Y.values[-1],'go')
plt.grid()
```

## Serie temporal

```python
temperaturas_paciente = 38 + 2*np.random.randn(24)
horas_medida          = pd.date_range('20180711',periods=24,freq='H')
ts = pd.Series(temperaturas_paciente, index=horas_medida)
print(ts)
```

```python
print(ts.index)
print(ts.dtype)
print(ts.values)
```

```python
print(horas_medida[23])
print(ts[23])
print(ts.iat[23])
print(ts.at[horas_medida[23]])

print(horas_medida[19:21])
print(ts[19:21])
print(ts.iloc[19:21])
print(ts.loc[horas_medida[19:21]])
```
