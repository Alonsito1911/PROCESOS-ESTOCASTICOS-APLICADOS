### 4º Frecuencia de aparición de los datos 

* El análisis estadístico de los datos implica estudiar cuantitativamente la repetición o frecuencia de los mismos. Ello se conoce como frecuencia absoluta de los datos. Además de la anterior, son de interés la frecuencia relativa al total, la frecuencia acumulada progresivamente y la frecuencia acumalada relativa. La repetición se aplica a datos tanto categóricos como discretos, o también a grupos de datos numéricos, ya sean continuos o discretos.
* Para estudiar datos categórico o discretos, como la localización de los clientes que aparencen en el conjunto de datos que teníamos en la variable *df_churn*, el análisis de la frecuencia de aparición se realiza ejecutando la siguiente instrucción:  ```LocalizacionFreq = pd.value_counts(df_churn['State'])```
* Muestre el resultado.
* Fijarse que nos muestra los acrónimos de los estados donde se localizan los clientes y sus frecuencias de aparición
* Fijarse que la lista está ordenada respecto a la frecuencia y no respecto a la variable.
* ¿Cual es el acrónimo del estado que tiene más clientes?¿Cuantos clientes tiene?
* Cuando los datos son numericos continuos o discretos en gran cantidad, como los gastos en la franja horaria del día, el análisis de la frecuencia absoluta de aparición se realiza ejecutando la siguiente instrucción: 
```gastoFrecAbs = pd.value_counts(gasto.iloc[:,0],bins=5,sort=False)```
* Muestre el resultado.
* Fijarse que los datos se han agrupado en 5 agrupamientos (bins=5) de igual tamaño respecto al valor de la variable, el gasto.
* Fijarse que el resultado no está ordenado respecto de la frecuencia (sort=False), sino respecto de la variable, o sea en cuanto a los bins o intervalos de los valores de gasto.
* ¿Que intervalo de valores de gasto es el más frecuente en la franja horaria de día? 
* ¿Cuantas veces se ha realizado un gasto entre 47.712 y 59.64 euros?

`````python
LocalizacionFreq = pd.value_counts(df_churn['State'])
print(LocalizacionFreq)
`````

`````python
gastoFrecAbs = pd.value_counts(gasto.iloc[:,0],bins=5,sort=False)
print(gastoFrecAbs)
`````

#### Más

```gastoFrecRel = pd.value_counts(gasto.iloc[:,0],bins=5,sort=False, normalize=True)```

* La frecuencia relativa acumulada nos da idea del porcentaje de la frecuencia acumulada respecto del total.
* Calcule las frecuencias relativas acumuladas al gasto para la franja horaria del día, agrupados en 5 intervalos iguales respecto a los valores de gasto, con la siguiente instrucción:
```gastoFrecRelAcc = gastoFrecRel.cumsum()```
* Fijarse que la suma acumulada lo hace sobre las frecuencias relativas del gasto que habíamos obtenido anteriormente.
